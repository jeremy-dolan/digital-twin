```python
## effort xhigh -- streaming and timestamps version

import time

t0 = time.perf_counter()
response = client.responses.create(
  model="gpt-5.2",
  input=[
    { "role": "developer", "content": [ { "type": "input_text", "text": PROMPT } ] },
    { "role": "user", "content": [ {
          "type": "input_text", "text": "So, then, what brings us to this mortal coil?"
    } ] },
  ],
  text={
    "format": { "type": "text" },
    "verbosity": "medium"
  },
  reasoning={
    "effort": "xhigh",
    "summary": "concise"
  },
  store=True,
  include=[
    "reasoning.encrypted_content",
    "web_search_call.action.sources"
  ]
  stream=True
)
print(f"+{time.perf_counter() - t0:.6f}s  ITERATION STARTS NOW")

for item in response:
    print(f"+{time.perf_counter() - t0:.6f}s  {item}")
```
