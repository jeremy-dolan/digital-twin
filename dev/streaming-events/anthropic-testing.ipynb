```python
import anthropic
client = anthropic.Anthropic()
```

```python
message = client.messages.create(
    model="claude-sonnet-4-6",
    system='For the given user question, output only an integer answer',
    messages=[
        {
            "role": "user",
            "content": "Please count the number of 'h's in the English spelling of the answer to 1298743*283764",
        }
    ],
    max_tokens=2_000,
    thinking={ "type": "adaptive" },
    output_config={ "effort": "medium" },  # reasoning effort
    stream=True,
)
```

```python
# now with .stream instead of .create

with client.messages.stream(
    model="claude-sonnet-4-6",
    system='For the given user question, output only an integer answer',
    messages=[
        {
            "role": "user",
            "content": "Please count the number of 'h's in the English spelling of the answer to 1298743*283764",
        }
    ],
    max_tokens=2_000,
    thinking={ "type": "adaptive" },
    output_config={ "effort": "medium" },  # reasoning effort
) as stream:
    for event in stream:
        print(event)
```

```python
stream.get_final_message()
```

```python
stream.get_final_message().model_dump()
```
